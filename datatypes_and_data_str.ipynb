19. Why are dictionaries considered unordered, and how does that affect data retrieval?
Answer:
In older versions of Python, dictionaries were unordered, which means the items didn’t stay in the same order as we added them.

From Python 3.7 onwards, dictionaries do keep the insertion order, but they are still not like lists — we can’t get items by their position (like dict[0]).

We always get values by using their keys, not by their place in the dictionary.

Example:
student_id= {"name": "Amit", "age": 20}
print(student_id["name"])  # ✅ Works
print(student_id[0])       # ❌ Error: Keys, not index


20. Explain in short the difference between a list and a dictionary in terms of data retrieval.
Answer:
In a list, we get data using an index number (like list[0]), but in a dictionary, we get data using a key (like dict["name"]).

List → position-based access

Dictionary → key-based access
